## Задание 4

---



* В таблице df_1 создайте новый столбец `SaleDiscount`.
* Пройдитесь по всему столбцу `HalfBath`, если в нем, в строке значение 0, то в столбце `SaleDiscount` значение одинаковое со столбцом `SalePrice`, если в `HalfBath` значение 1, то в `SaleDiscount` значение на 10 000 ниже чем в `SalePrice`, если в `HalfBath` значение 2, то в `SaleDiscount` значение на 10 000 выше, чем в `SalePrice`.

В результате у вас должно получиться так:

image.png


image.png

```python
# Решение

df_1['SaleDiscount'] = df_1['SalePrice']
# Условие
df_1.loc[df_1['HalfBath'] == 1, 'SaleDiscount'] = df_1['SalePrice'] - 10000
df_1.loc[df_1['HalfBath'] == 2, 'SaleDiscount'] = df_1['SalePrice'] + 10000
# Фильтуем
df_halfbath_0 = df_1[df_1['HalfBath'] == 0][['HalfBath', 'SalePrice', 'SaleDiscount']]
df_halfbath_1 = df_1[df_1['HalfBath'] == 1][['HalfBath', 'SalePrice', 'SaleDiscount']]
df_halfbath_2 = df_1[df_1['HalfBath'] == 2][['HalfBath', 'SalePrice', 'SaleDiscount']]

display(df_halfbath_0.head())
display(df_halfbath_1.head())
display(df_halfbath_2.head())
```

Файл excel, как известно может содержать несколько листов в одной таблице.

Пример:\

```python
path = 'https://docs.google.com/spreadsheets/d/1HwcQN30MATceqBf5z9Bc8BsZ1J0oSyLB/edit?usp=drive_link&ouid=109837745230231498702&rtpof=true&sd=true'
```

```python
path_2 = "https://docs.google.com/spreadsheets/d/1HwcQN30MATceqBf5z9Bc8BsZ1J0oSyLB/export?format=xlsx"
```

```python
# Решение

xls = pd.ExcelFile(path_2, engine='openpyxl')

df1 = xls.parse('Лист1')
df2 = xls.parse('Лист2')

# Удалем первую строку и делаем вторую строку в качестве заголовка
df1 = df1.dropna(how='all').reset_index(drop=True)
df1.columns = df1.iloc[0]
df1 = df1.drop(0).reset_index(drop=True)

# Удалем первую строку и делаем вторую строку в качестве заголовка
df2 = df2.dropna(how='all').reset_index(drop=True)
df2.columns = df2.iloc[0]
df2 = df2.drop(0).reset_index(drop=True)


display(df1)
display(df2)
```
