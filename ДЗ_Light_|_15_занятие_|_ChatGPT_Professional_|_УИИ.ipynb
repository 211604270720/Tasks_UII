```python
from IPython.display import Audio
import whisper
from pprint import pprint
import matplotlib.pyplot as plt
import time
```

```python
Audio("/content/debat.mp3")
```

```python
start_time = time.time()
model = whisper.load_model("tiny")
result_tiny = model.transcribe("/content/debat.mp3")
end_time = time.time()

print(result_tiny['text'])
print(" ")
tiny_time = end_time - start_time
print("Время выполнения: ", tiny_time, "секунд")
```

```python
# Значения для столбцов
values = [tiny_time, base_time, small_time, medium_time, large_time]

# Названия для столбцов
categories = ['tiny', 'base', 'small', 'medium', 'large']

# Создание столбчатой диаграммы
plt.bar(categories, values)

# Добавление названия диаграммы и осей
plt.title("Сравнение скорости распознования моделей whisper")
plt.xlabel("Модели")
plt.ylabel("Секунды")

# Отображение диаграммы
plt.show()
```
