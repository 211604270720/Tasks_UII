## Задание 3

---

* Создай таблицу товаров из магазина электроники, назовите её `df3`. Должно быть 5 строк и 3 столбца: `['Наименование товара', 'Цена', 'Остаток']`.\
`shape(5, 3)`
* После того, как создали таблицу добавьте ещё одну строку с умной колонкой.\
 `shape(6, 3)`

Результат выведите на экран.

```python
# Решение

df_e = {
    'Наименование': ['Телевизор', 'Наушники', 'Планшет', 'Компьютер', 'Монитор'],
    'Цена': [55900, 2300, 17800, 78000, 14300],
    'Остаток': [12, 21, 26, 19, 15]
}

df3 = pd.DataFrame(df_e)
df3
```

```python
import warnings
warnings.filterwarnings('ignore')

df3 = df3.append({'Наименование': 'Умная колонка', 'Цена': 7900, 'Остаток': 30}, ignore_index=True)
df3
```

## Задача 4

---

Отсортируйте получившуюся таблицу `df3`, по столбцу "цена". При этом индексы должны идти по порядку. (По убыванию)

```python
# Решение

df3 = df3.sort_values(by="Цена", ascending=False).reset_index(drop=True)
df3
```
