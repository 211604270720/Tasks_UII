### Задание 2

---



* разделите текст по предложениям и присвойте каждое предложение новой переменной.
* сделайте вывод так, чтобы каждое предложение было с новой строки, в конце предложения точка.

Пример:
```
Скороговорка — это короткая, синтаксически правильная фраза на любом языке с искусственно усложнённой артикуляцией.
Скороговорки содержат близкие по звучанию, но различные фонемы и сложные для произношения сочетания фонем.
Зачастую содержат аллитерации и рифмы.
```

````python
patter = '''Скороговорка — это короткая,
синтаксически правильная фраза на любом языке
с искусственно усложнённой артикуляцией.
Скороговорки содержат близкие по звучанию,
но различные фонемы и сложные
для произношения сочетания фонем.
Зачастую содержат аллитерации и рифмы.'''

print(patter)
````

````python
# Решение

# Разделяем на предложения
text = re.split(r'(?<=[.!?])\s+', patter.replace('\n', ' ').strip())

# Присваиваем каждому предложению переменную
per_dic = {}
for i, text in enumerate(text, 1):
    per_dic[f'text_{i}'] = text

# Вывод предложений
out = "\n".join([per_dic[key].rstrip('.') + '.' for key in per_dic])
print(out)
````

### Задание 3

Напишите функцию, которая выбирает 2-х значные числа из списка с помощью функции filter и возвращает результат новым списком.

````python
lst = [96, 186, 11, 116, 183,  95, 181, 95, 55, 128, 54, 24, 21, 198, 118, 46, 166, 137, 24]
````

````python
# Решение

def num(lst):
    return list(filter(lambda x: 10 <= x < 100, lst))

num(lst)
````

### Задание 4

Дан текст, с помощью регулярных выражений выведите только цену.\
Пример:


image.png

````python
text = '''
Товар: Яблоки, Цена: 2.99 USD
Товар: Молоко, Цена: 1.50 USD
Товар: Шоколад, Цена: 3.25 USD
Товар: Кофе, Цена: 5.99 USD
Товар: Мороженое, Цена: 4.75 USD
'''

# Решение

text_1 = re.findall(r'Цена: (\d+\.\d+)', text)
prices = ['Цена: ' + price + ' USD' for price in text_1]
output = '\n'.join(prices)
print(output)
````

### Задание 5

---



Ваш код принимает на вход строку string, которая содержит разные символы: буквы, цифры, знаки препинания и т.д. Напишите функцию, которая принимает строку и возвращает 3 списка, список со всеми буквами в верхнем регистре, список в нижнем регистре, список только с цифрами. 3 списка должны находится в словаре.
Пример:



image.png

````python
# Решение

def str_gen(s):
    # Генераторы списков для цифр, заглавных и строчных букв
    num = [i for i in s if i.isdigit()]
    up = [i for i in s if i.isupper()]
    low = [i for i in s if i.islower()]

    return {
        'Все цифры': num,
        'Все заглавные': up,
        'Все строчные': low
    }
string = 'jljdfsjIJIsdjjkdljfiIWJe878&asd876dsdyu&Asd8776as68'

# Применяем функцию к строке
result_gen = str_gen(string)
f_result_gen = (
    f"Все цифры: {result_gen['Все цифры']}\n"
    f"Все заглавные: {result_gen['Все заглавные']}\n"
    f"Все строчные: {result_gen['Все строчные']}"
)
print(f_result_gen)
````
