### Задание 1

---



* Создайте текстовый файл с именем "Пушкин.txt".
* Запишите в него из стихотворение A.C. Пушкин все слова начинающиеся на букву "П", "п" - используйте регулярные выражения.
* Прочитайте получившийся файл


Пример:

image.png

```python
import re
import os

Pushkin = '''
Октябрь уж наступил — уж роща отряхает
Последние листы с нагих своих ветвей;
Дохнул осенний хлад — дорога промерзает.
Журча еще бежит за мельницу ручей,
Но пруд уже застыл; сосед мой поспешает
В отъезжие поля с охотою своей,
И страждут озими от бешеной забавы,
И будит лай собак уснувшие дубравы.
'''
```

```python
# Решение

# Извлекаем слова начинающиеся на букву "П" или "п"
lit = re.findall(r"\b[Пп]\w+", Pushkin)

# Запись слов в файл
with open("/content/Пушкин.txt", 'w', encoding="utf-8") as file:
    for word in lit:
        file.write(word + "\n")

# Чтение из файла
with open("/content/Пушкин.txt", 'r', encoding="utf-8") as file:
    push_lit = file.read()

print(push_lit)
```

### Задание 3

---



Напишите функцию, которая читает любой файл `.txt` и выводит словарь с длинной слов.

image.png

```python
# Решение

def len_w(filename):
    with open(filename, 'r', encoding="utf-8") as file:
        con = file.read()

    # Разбиваем текст на слова
    words = con.split()

    # Создаем словарь слов по длинам
    leng = {}
    for word in words:
        len_word = len(word)
        if len_word not in leng:
            leng[len_word] = []
        leng[len_word].append(word)

    return leng

dict_1 = len_w("/content/Пушкин.txt")

out = ""
for leng, words in dict_1.items():
    out += f"Длинна {leng} букв: {words}\n"

print(out)
```

### Задание 4

---



Дан список text, состоящий из подстрок вида "буква число". Напишите программу на Python, которая отсортирует этот список по убыванию числа и выведет на экран отсортированный список. Используйте встроенный **sorted** для соритровки.\
\
Пример:\
`['г 296', 'и 45', 'я 22', 'б 12', 'ш 10', 'а 5']`

```python
# Решение

text = ['ш 10','и 45','г 296', 'я 22', 'а 5','б 12']

# Сортировка списка по убыванию числа
sort_text = sorted(text, key=lambda x: -int(x.split()[1]))

sort_text
```

```python
# Решение

# Создаем функуцию
def s_dict(d, ascending=True):
    s_items = sorted(d.items(), key=lambda x: x[1], reverse=not ascending)
    return dict(s_items)

my_dict = {'a':500, 'b':5874, 'c': 560,'d':400, 'e':5874, 'f': 20}
s_dict(my_dict, ascending=True)
```
